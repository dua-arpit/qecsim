```python
import numpy as np

L=13

def map(i,j,k,L):
    return int(i+j*L+k*L*L)

rmnodes1=np.array([],dtype=int)
rmnodes2=np.array([],dtype=int)
rmnodes3=np.array([],dtype=int)

if L in range(3,9):
    x=int(np.round(L/3))
    iv1=x
    
    for i in range(iv1,iv1+x):
        for j in range(iv1,iv1+x):
            for k in range(iv1,iv1+x):
                rmnodes1=np.append(rmnodes1,map(i,j,k,L))   
                
if L in range(9,27):
    x=int(np.round(L/3))
    y=int(np.round(x/3))
    y2=int(np.round((L-2*x)/3))
    iv1=x
    iv2=[y,x+y,2*x+y2] 
    
    for i in range(iv1,iv1+x):
        for j in range(iv1,iv1+x):
            for k in range(iv1,iv1+x):
                rmnodes1=np.append(rmnodes1,map(i,j,k,L))

    corners=np.array([[0,0,0]],dtype=int)
    for xi in iv2:
        for xj in iv2:
            for xk in iv2:
                if map(xi,xj,xk,L) not in rmnodes1:
                    corners=np.concatenate((corners,[[xi,xj,xk]]),axis=0)
    corners=np.delete(corners,0,0)   
    
    for corner in corners:
        iv,jv,kv=corner
        if iv!=2*x and jv!=2*x and kv!=2*x:
            for i in range(iv,iv+y):
                for j in range(jv,jv+y):
                    for k in range(kv,kv+y):
                        if map(i,j,k,L) not in rmnodes1:
                            rmnodes2=np.append(rmnodes2,map(i,j,k,L))
        else:
            for i in range(iv,iv+y2):
                for j in range(jv,jv+y2):
                    for k in range(kv,kv+y2):
                        if map(i,j,k,L) not in rmnodes1:
                            rmnodes2=np.append(rmnodes2,map(i,j,k,L))  


if L in range(27,37):
    x=int(np.round(L/3))
    y=int(np.round(x/3))
    y2=int(np.round((L-2*x)/3))
    z=int(np.round(y/3))
    z2=int(np.round((x-2*y)/3))
    iv1=x
    iv2=[y,x+y,2*x+y2]
    zr=int(np.round(y2/3))
    zr2=int(np.round((L-2*x-2*y2)/3))
    iv3=[z,y+z,2*y+z2, x+z,x+y+z,x+2*y+z2,2*x+zr,2*x+y2+zr,2*x+2*y2+zr2]

    for i in range(iv1,iv1+x):
        for j in range(iv1,iv1+x):
            for k in range(iv1,iv1+x):
                rmnodes1=np.append(rmnodes1,map(i,j,k,L))

    corners=[[0,0,0]]
    for xi in iv2:
        for xj in iv2:
            for xk in iv2:
                if map(xi,xj,xk,L) not in rmnodes1:
                    corners=np.concatenate((corners,[[xi,xj,xk]]),axis=0)
    corners=np.delete(corners,0,0)   

    for corner in corners:
        iv,jv,kv=corner
        if iv!=2*x and jv!=2*x and kv!=2*x:
            for i in range(iv,iv+y):
                for j in range(jv,jv+y):
                    for k in range(kv,kv+y):
                        if map(i,j,k,L) not in rmnodes1:
                            rmnodes2=np.append(rmnodes2,map(i,j,k,L))
        else:
            for i in range(iv,iv+y2):
                for j in range(jv,jv+y2):
                    for k in range(kv,kv+y2):
                        if map(i,j,k,L) not in rmnodes1:
                            rmnodes2=np.append(rmnodes2,map(i,j,k,L)) 


    corners2=[[0,0,0]]
    for xi in iv3:
        for xj in iv3:
            for xk in iv3:
                if map(xi,xj,xk,L) not in rmnodes1 and map(xi,xj,xk,L) not in rmnodes2:
                    corners2=np.concatenate((corners2,[[xi,xj,xk]]),axis=0)
    corners2=np.delete(corners2,0,0)      

    for corner in corners2:
        iv,jv,kv=corner        
        rmnodes3=np.append(rmnodes3,map(iv,jv,kv,L))                            
```

```python
def hole_nodes(L,level):
    rmnodes1=np.array([],dtype=int)
    rmnodes2=np.array([],dtype=int)
    rmnodes3=np.array([],dtype=int)

    if L in range(3,9):
        x=int(np.round(L/3))
        iv1=x
        
        for i in range(iv1,iv1+x):
            for j in range(iv1,iv1+x):
                for k in range(iv1,iv1+x):
                    rmnodes1=np.append(rmnodes1,map(i,j,k,L))   
                    
    if L in range(9,27):
        x=int(np.round(L/3))
        y=int(np.round(x/3))
        y2=int(np.round((L-2*x)/3))
        iv1=x
        iv2=[y,x+y,2*x+y2] 
        
        rmnodes_coord=[[0,0,0]]
        for i in range(iv1,iv1+x):
            for j in range(iv1,iv1+x):
                for k in range(iv1,iv1+x):
                    rmnodes1=np.append(rmnodes1,map(i,j,k,L))
                    rmnodes_coord=np.concatenate((rmnodes_coord,[[i,j,k]]),axis=0)
        rmnodes_coord=np.delete(rmnodes_coord,0,0)

        corners=[[0,0,0]]
        for xi in iv2:
            for xj in iv2:
                for xk in iv2:
                    if map(xi,xj,xk,L) not in rmnodes1:
                        corners=np.concatenate((corners,[[xi,xj,xk]]),axis=0)
        corners=np.delete(corners,0,0)   

        for corner in corners:
            iv,jv,kv=corner
            if iv!=2*x and jv!=2*x and kv!=2*x:
                for i in range(iv,iv+y):
                    for j in range(jv,jv+y):
                        for k in range(kv,kv+y):
                            if map(i,j,k,L) not in rmnodes1:
                                rmnodes2=np.append(rmnodes2,map(i,j,k,L))
                                rmnodes_coord=np.concatenate((rmnodes_coord,[[i,j,k]]),axis=0)

            else:
                for i in range(iv,iv+y2):
                    for j in range(jv,jv+y2):
                        for k in range(kv,kv+y2):
                            if map(i,j,k,L) not in rmnodes1:
                                rmnodes2=np.append(rmnodes2,map(i,j,k,L)) 
                                rmnodes_coord=np.concatenate((rmnodes_coord,[[i,j,k]]),axis=0)


    if L in range(27,37):
        x=int(np.round(L/3))
        y=int(np.round(x/3))
        y2=int(np.round((L-2*x)/3))
        z=int(np.round(y/3))
        z2=int(np.round((x-2*y)/3))
        iv1=x
        iv2=[y,x+y,2*x+y2]
        zr=int(np.round(y2/3))
        zr2=int(np.round((L-2*x-2*y2)/3))
        iv3=[z,y+z,2*y+z2, x+z,x+y+z,x+2*y+z2,2*x+zr,2*x+y2+zr,2*x+2*y2+zr2]

        rmnodes_coord=[[0,0,0]]
        for i in range(iv1,iv1+x):
            for j in range(iv1,iv1+x):
                for k in range(iv1,iv1+x):
                    rmnodes1=np.append(rmnodes1,map(i,j,k,L))
                    rmnodes_coord=np.concatenate((rmnodes_coord,[[i,j,k]]),axis=0)
        rmnodes_coord=np.delete(rmnodes_coord,0,0)

        corners=[[0,0,0]]
        for xi in iv2:
            for xj in iv2:
                for xk in iv2:
                    if map(xi,xj,xk,L) not in rmnodes1:
                        corners=np.concatenate((corners,[[xi,xj,xk]]),axis=0)
        corners=np.delete(corners,0,0)   

        for corner in corners:
            iv,jv,kv=corner
            if iv!=2*x and jv!=2*x and kv!=2*x:
                for i in range(iv,iv+y):
                    for j in range(jv,jv+y):
                        for k in range(kv,kv+y):
                            if map(i,j,k,L) not in rmnodes1:
                                rmnodes2=np.append(rmnodes2,map(i,j,k,L))
                                rmnodes_coord=np.concatenate((rmnodes_coord,[[i,j,k]]),axis=0)

            else:
                for i in range(iv,iv+y2):
                    for j in range(jv,jv+y2):
                        for k in range(kv,kv+y2):
                            if map(i,j,k,L) not in rmnodes1:
                                rmnodes2=np.append(rmnodes2,map(i,j,k,L)) 
                                rmnodes_coord=np.concatenate((rmnodes_coord,[[i,j,k]]),axis=0)

        corners2=[[0,0,0]]
        for xi in iv3:
            for xj in iv3:
                for xk in iv3:
                    if map(xi,xj,xk,L) not in rmnodes1 and map(xi,xj,xk,L) not in rmnodes2:
                        corners2=np.concatenate((corners2,[[xi,xj,xk]]),axis=0)
        corners2=np.delete(corners2,0,0)      

        for corner in corners2:
            i,j,k=corner        
            rmnodes3=np.append(rmnodes3,map(i,j,k,L))
            rmnodes_coord=np.append(rmnodes_coord,[[i,j,k]])


    if level==1:
        rmnodes=rmnodes1
    elif level==2:
        rmnodes=np.union1d(rmnodes1,rmnodes2)
    elif level==3:
        reduce(np.union1d,(rmnodes1,rmnodes2,rmnodes3))
    
    return rmnodes,rmnodes_coord


rmnodes,rmnodes_coord=hole_nodes(9,2)
```
